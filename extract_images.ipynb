```python
#import libraries 
import pydicom
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#resize the image

# img = np.transpose(img, (1, 2, 0 ))
dcm =  pydicom.dcmread(rf"C:\Users\Wahaj Sajid\Desktop\Research\MIDRC\Asian Female Positive COVID19 (Age 0 - 75)\2.16.840.1.114274.1818.46526962804395084737001273829094365887.dcm")
img =  dcm.pixel_array#access the pixel data

# If the image is not 2D, try transposing (common for segmentation masks)
if img.ndim == 3 and img.shape[0] <= 4:  # Likely (channels, rows, columns)
    img = np.transpose(img, (1, 2, 0))

# If the image is still not displayable, try displaying a single channel
if img.ndim == 3:
    img = img[..., 0]  # Take first channel

# Optional: Apply scaling if needed
if hasattr(dcm, 'RescaleSlope') and hasattr(dcm, 'RescaleIntercept'):
    img = img * dcm.RescaleSlope + dcm.RescaleIntercept

# Display the image
plt.imshow(img, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```
